```python
import numpy as np
import pandas as pd
c=pd.read_csv("50_Startups.csv")
c
```

```python
data=c["Profit"]
data
```

```python
from scipy.stats import skew
ans=skew(data)
if ans>0:
    print("positive skew")
elif ans<0:
    print("negative skew")
else:
    print("normal skew")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.hist(data, color='skyblue', bins=10)
plt.title("Profit Distribution")
plt.xlabel("Profit")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
