```python
#if the call center received 10 calls per min 
#means,how much of minutes need for receiving 
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson as p
lam=10
k=15
pos=p.pmf(k,lam)
print("probability of getting 15 call in a minute:",pos)
x=np.arange(0,21)
fit=p.pmf(x,lam)
plt.bar(x,fit,color="purple")
plt.title("poisson distribution")
plt.xlabel("no of calls in per minute")
plt.ylabel("probability")#chances 
plt.show()
```
